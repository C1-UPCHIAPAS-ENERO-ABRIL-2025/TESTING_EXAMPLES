# HOW TO: Configurando el entorno para nuestras primeras pruebas estáticas.

El análisis estático es una técnica que requiere herramientas específicas que inspeccionan el código sin ejecutarlo. El análisis depende mucho de la herramienta que se use, en general se buscan errores de tipado o inconsistencias, vulnerabilidades y exposición de secretos, complejidad ciclomática, cumplimiento de infraestructuras de código, entre otros. En esta ocasión nos centraremos en 4 herramientas de análisis estático que nos ayudarán a establecer un estándar de calidad de código y a detectar problemas comunes en nuestros proyectos. Estas herramientas son:

- **Pylint**: para análisis de estilo y calidad de código en Python.
- **mypy**: para análisis de tipos en Python.
- **Radon**: para análisis de complejidad ciclomática en Python.
- **Bandit**: para análisis de seguridad en Python.
- **Hadolint**: para análisis de seguridad de Dockerfiles.
- **GPG**: para verificación de firmas digitales en reportes.

## Estandarizando el análisis estático

No se puede analizar nada si no se tiene una métrica con que comparar los resultados. Cada herramienta se especializa en un campo pero si no sabes que significa cada métrica o que es un resultado aceptable, entonces no podrás sacar el máximo provecho de ellas. Lo que queremos hacer se divide en 2 fases muy claras: 

1. **Definir métricas de calidad aceptables** para cada herramienta. Las métricas de calidad establecen cuanto se considera un código de buena calidad. Estas métricas pueden variar dependiendo del proyecto y del equipo, pero lo importante es que sean claras, reales, alcanzables y medibles de manera objetiva y que no dependan del sistema operativo, del entorno de desarrollo, la configuración de la herramienta o cualquier otro factor externo. Lo esencial es que sean reproducibles sin importar qué herramienta se utilice para medirlas.

2. **Saber en qué se basan los estándares de medición**. Cada herramienta tiene sus propios estándares de medición, es decir, cada una utiliza una base o metodología específica para construir sus métricas.

	<ol type="a">
		<li><strong>Pylint</strong>: se basa en PEP 8 (la guía de estilo oficial de Python) y en convenciones de código establecidas por la comunidad Python. Evalúa el código contra estas reglas para determinar su adherencia a estándares de calidad.</li>
		<li><strong>mypy</strong>: se basa en el sistema de tipos de Python definido en PEP 484 y utiliza inferencia estática de tipos para analizar anotaciones de tipos. Mide qué porcentaje del código tiene cobertura de tipos explícita.</li>
		<li><strong>Radon</strong>: se basa en la teoría de complejidad ciclomática de McCabe, que analiza el grafo de flujo de control del código. Cuenta el número de caminos independientes que existen en una función o método.</li>
		<li><strong>Bandit</strong>: se basa en un catálogo de patrones de código inseguro y vulnerabilidades conocidas. Utiliza análisis del árbol de sintaxis abstracta (AST) para detectar construcciones que podrían ser explotadas.</li>
		<li><strong>Hadolint</strong>: se basa en las reglas oficiales de buenas prácticas de Docker. Evalúa los Dockerfiles contra estándares de seguridad y eficiencia establecidos por la comunidad Docker.</li>
		<li><strong>GPG</strong>: se basa en criptografía de clave asimétrica. Utiliza una clave privada para firmar reportes y una clave pública para verificar su autenticidad e integridad, garantizando que no han sido alterados.</li>
	</ol>

## Conociendo las métricas de análisis estático

### Linting

Un linter es una herramienta que analiza el código fuente para detectar errores de sintaxis, problemas de estilo y otras cosas que tengan que ver con no seguir una convención estándar de código. Pylint nos ayuda a verificar que esto se cumple ya que se basa en PEP 8, la guía de estilo para Python. Pylint asigna una puntuación de calidad al código, que va de 0 a 10, donde 10 es el mejor resultado posible.

### Tipado

El tipado es una técnica que se utiliza para especificar el tipo de datos que se espera en una variable, función o clase. Si bien python es un lenguaje de tipado dinámico, es posible utilizar anotaciones de tipos para mejorar la legibilidad del código y detectar errores de tipo en tiempo de desarrollo. No es que python no pueda manejar tipados dinámicos, es que el tipado estático nos ayuda a detectar errores de tipo antes de ejecutar el código, lo que puede ahorrar tiempo y esfuerzo en la depuración. Mypy es una herramienta que nos ayuda a verificar que el código tiene anotaciones de tipos completas y correctas, y nos asigna un porcentaje de cobertura de tipos, donde 100% es el mejor resultado posible.

### Complejidad ciclomática

La complejidad ciclomática es una métrica que se utiliza para medir la complejidad de un código. La complejidad de un código se mide según el número de caminos independientes que existen en el código. Una función puede tener varios caminos independientes, estos caminos se generan por ramas condicionales, bucles y excepciones. Cuantos más caminos independientes tenga una función, más compleja será. Radon es una herramienta que nos ayuda a medir la complejidad ciclomática de nuestro código, y nos asigna una puntuación de complejidad, donde A es el mejor resultado posible (complejidad ciclomática menor a 10) y F es el peor resultado posible (complejidad ciclomática mayor a 50).

### Seguridad

La seguridad es una preocupación importante en el desarrollo de software, ya que un código inseguro puede ser explotado por atacantes y robar datos o secuestrar el desarrollo para pedir un rescate. Bandit es una herramienta que nos ayuda a detectar vulnerabilidades de seguridad en nuestro código, y nos asigna una puntuación de seguridad, donde 0 es el mejor resultado posible (sin vulnerabilidades críticas) y 10 es el peor resultado posible (con vulnerabilidades críticas). Hadolint es una herramienta similar pero se enfoca en detectar vulnerabilidades de seguridad en los Dockerfiles.

### Verificación de firmas digitales

La verificación de firmas digitales es una técnica que se utiliza para garantizar la autenticidad e integridad de los reportes generados por las otras herramientas. GPG es una herramienta que nos ayuda a verificar la autenticidad e integridad de los reportes, y nos asigna un resultado de verificación, donde "Good signature" es el mejor resultado posible (reporte firmado y verificado) y "Bad signature" es el peor resultado posible (reporte no firmado o con firma inválida).

### Pylint - Análisis de Calidad de Código

Esta celda ejecuta Pylint sobre ambos proyectos y:
1. Genera reportes de calidad de código (`reports_pylint_vulnerable.txt` y `reports_pylint_clean.txt`)
2. Firma digitalmente cada reporte con GPG (genera archivos `.sig`)
3. Muestra un resumen de las puntuaciones obtenidas

**Métricas a revisar:**
- Puntuación total (0-10)
- Errores de convención (C)
- Errores de refactorización (R)
- Warnings (W)
- Errores (E)

````python
import subprocess

print("=" * 70)
print("PYLINT - Code Quality Analysis")
print("=" * 70)

projects = {
    "vulnerable": VULNERABLE_PROJECT,
    "clean": CLEAN_PROJECT
}

for project_type, project_name in projects.items():
    report_file = REPORTS_DIR / f"pylint_{project_type}.txt"
    sig_file = REPORTS_DIR / f"pylint_{project_type}.txt.sig"
    
    print(f"\n{'-' * 70}")
    print(f"Analyzing: {project_name}")
    print(f"{'-' * 70}")
    
    # Run pylint
    cmd = [str(VENV_PYTHON), "-m", "pylint", f"{project_name}/inventory_system"]
    
    try:
        result = subprocess.run(cmd, capture_output=True, text=True, timeout=60)
        
        # Save report
        with open(str(report_file), "w", encoding="utf-8") as f:
            f.write(result.stdout)
            if result.stderr:
                f.write("\n\n=== STDERR ===\n")
                f.write(result.stderr)
        
        print(f"OK - Report saved: {report_file.name}")
        
        # Extract score
        for line in result.stdout.split("\n"):
            if "Your code has been rated" in line:
                print(f"  Score: {line.strip()}")
                break
        
        # Sign report with GPG (no passphrase)
        if GPG_KEY_ID != "NO_GPG_KEY":
            sign_cmd = [
                "gpg",
                "--default-key", GPG_KEY_ID,
                "--output", str(sig_file),
                "--detach-sign",
                str(report_file)
            ]
            
            sign_result = subprocess.run(sign_cmd, capture_output=True, text=True, timeout=30)
            
            if sign_result.returncode == 0:
                print(f"OK - Signature saved: {sig_file.name}")
            else:
                print(f"WARNING - Could not sign report: {sign_result.stderr[:60]}")
        else:
            print("WARNING - Signature skipped (no GPG key configured)")
    
    except subprocess.TimeoutExpired:
        print(f"ERROR - Analysis timed out for {project_name}")
    except Exception as e:
        print(f"ERROR - {str(e)}")

print(f"\n{'=' * 70}")
````

### Bandit - Análisis de Seguridad

Esta celda ejecuta Bandit sobre ambos proyectos y:
1. Genera reportes de seguridad en texto plano (`reports_bandit_vulnerable.txt` y `reports_bandit_clean.txt`)
2. Genera reportes adicionales en formato JSON para análisis programático
3. Firma digitalmente cada reporte de texto con GPG (genera archivos `.sig`)
4. Muestra un resumen de vulnerabilidades encontradas

**Métricas a revisar:**
- Severidad de issues: LOW, MEDIUM, HIGH
- Confianza: LOW, MEDIUM, HIGH
- Patrones detectados: hardcoded passwords, uso inseguro de pickle, ejecución de comandos shell, etc.

````python
import subprocess
import json

print("=" * 70)
print("BANDIT - Security Analysis")
print("=" * 70)

projects = {
    "vulnerable": VULNERABLE_PROJECT,
    "clean": CLEAN_PROJECT
}

for project_type, project_name in projects.items():
    report_txt_file = REPORTS_DIR / f"bandit_{project_type}.txt"
    report_json_file = REPORTS_DIR / f"bandit_{project_type}.json"
    sig_file = REPORTS_DIR / f"bandit_{project_type}.txt.sig"
    
    print(f"\n{'-' * 70}")
    print(f"Analyzing: {project_name}")
    print(f"{'-' * 70}")
    
    # Run bandit text report
    cmd_txt = [str(VENV_PYTHON), "-m", "bandit", "-r", f"{project_name}/inventory_system", "-ll"]
    
    try:
        result_txt = subprocess.run(cmd_txt, capture_output=True, text=True, timeout=60)
        
        # Save text report
        with open(str(report_txt_file), "w", encoding="utf-8") as f:
            f.write(result_txt.stdout)
            if result_txt.stderr:
                f.write("\n\n=== STDERR ===\n")
                f.write(result_txt.stderr)
        
        print(f"OK - Text report saved: {report_txt_file.name}")
        
        # Run bandit JSON report
        cmd_json = [str(VENV_PYTHON), "-m", "bandit", "-r", f"{project_name}/inventory_system", 
                    "-f", "json", "-o", str(report_json_file)]
        
        result_json = subprocess.run(cmd_json, capture_output=True, text=True, timeout=60)
        
        if result_json.returncode in [0, 1]:
            print(f"OK - JSON report saved: {report_json_file.name}")
            
            # Analyze JSON
            try:
                with open(str(report_json_file), "r", encoding="utf-8") as f:
                    bandit_data = json.load(f)
                    
                issue_count = len(bandit_data.get("results", []))
                
                if issue_count == 0:
                    print("STATUS: OK - No vulnerabilities found")
                else:
                    high = sum(1 for r in bandit_data["results"] if r["issue_severity"] == "HIGH")
                    medium = sum(1 for r in bandit_data["results"] if r["issue_severity"] == "MEDIUM")
                    low = sum(1 for r in bandit_data["results"] if r["issue_severity"] == "LOW")
                    
                    print(f"STATUS: WARNING - {issue_count} vulnerabilities found")
                    print(f"  HIGH: {high}, MEDIUM: {medium}, LOW: {low}")
            except Exception as e:
                print(f"WARNING - Could not analyze JSON: {str(e)}")
        
        # Sign report with GPG (no passphrase)
        if GPG_KEY_ID != "NO_GPG_KEY":
            sign_cmd = [
                "gpg",
                "--default-key", GPG_KEY_ID,
                "--output", str(sig_file),
                "--detach-sign",
                str(report_txt_file)
            ]
            
            sign_result = subprocess.run(sign_cmd, capture_output=True, text=True, timeout=30)
            
            if sign_result.returncode == 0:
                print(f"OK - Signature saved: {sig_file.name}")
            else:
                print(f"WARNING - Could not sign report: {sign_result.stderr[:60]}")
        else:
            print("WARNING - Signature skipped (no GPG key configured)")
    
    except subprocess.TimeoutExpired:
        print(f"ERROR - Analysis timed out for {project_name}")
    except Exception as e:
        print(f"ERROR - {str(e)}")

print(f"\n{'=' * 70}")
````

### Verificación de Firmas Digitales

Esta celda verifica todas las firmas digitales generadas en las celdas anteriores para garantizar:
1. **Autenticidad**: Los reportes fueron firmados por la clave GPG configurada
2. **Integridad**: Los reportes no han sido modificados después de ser firmados

**Qué revisar:**
- `Good signature from "Tu Nombre <tu.email@example.com>"` = OK - Firma válida
- `BAD signature` = ERROR - Archivo modificado o firma corrupta
- Advertencias de confianza son normales si no has establecido el web of trust

````python
import subprocess
from pathlib import Path

print("=" * 70)
print("DIGITAL SIGNATURE VERIFICATION")
print("=" * 70)

if GPG_KEY_ID == "NO_GPG_KEY":
    print("\nWARNING: GPG_KEY_ID not configured. No signatures to verify.")
else:
    tools = ["pylint", "mypy", "radon", "bandit"]
    project_types = ["vulnerable", "clean"]
    
    verification_results = []
    
    for tool in tools:
        for project_type in project_types:
            report_file = REPORTS_DIR / f"{tool}_{project_type}.txt"
            sig_file = REPORTS_DIR / f"{tool}_{project_type}.txt.sig"
            
            # Check if files exist
            if not report_file.exists():
                verification_results.append({
                    "tool": tool,
                    "project": project_type,
                    "status": "WARNING - Report not found",
                    "file": report_file
                })
                continue
            
            if not sig_file.exists():
                verification_results.append({
                    "tool": tool,
                    "project": project_type,
                    "status": "WARNING - Signature not found",
                    "file": sig_file
                })
                continue
            
            # Verify signature
            verify_cmd = ["gpg", "--verify", str(sig_file), str(report_file)]
            
            try:
                result = subprocess.run(verify_cmd, capture_output=True, text=True, timeout=30)
                
                output = result.stderr
                
                if "Good signature" in output:
                    signer = "Unknown user"
                    for line in output.split("\n"):
                        if "Good signature from" in line:
                            signer = line.split("Good signature from")[1].strip().strip('"')
                            break
                    
                    verification_results.append({
                        "tool": tool,
                        "project": project_type,
                        "status": "OK - Valid signature",
                        "signer": signer,
                        "file": report_file
                    })
                elif "BAD signature" in output:
                    verification_results.append({
                        "tool": tool,
                        "project": project_type,
                        "status": "ERROR - Invalid signature",
                        "file": report_file
                    })
                else:
                    verification_results.append({
                        "tool": tool,
                        "project": project_type,
                        "status": "ERROR - Verification error",
                        "file": report_file
                    })
            
            except subprocess.TimeoutExpired:
                verification_results.append({
                    "tool": tool,
                    "project": project_type,
                    "status": "TIMEOUT",
                    "file": report_file
                })
            except Exception as e:
                verification_results.append({
                    "tool": tool,
                    "project": project_type,
                    "status": f"ERROR: {str(e)[:30]}",
                    "file": report_file
                })
    
    # Display results
    print(f"\n{'-' * 70}")
    print("VERIFICATION RESULTS")
    print(f"{'-' * 70}\n")
    
    for result in verification_results:
        print(f"{result['status']:25} | {result['tool']:8} | {result['project']:10}")
        if "signer" in result:
            print(f"{'':25} | Signed by: {result['signer']}")
    
    # Summary
    valid = sum(1 for r in verification_results if "OK" in r["status"])
    invalid = sum(1 for r in verification_results if "ERROR" in r["status"])
    missing = sum(1 for r in verification_results if "WARNING" in r["status"])
    
    print(f"\n{'-' * 70}")
    print(f"SUMMARY: {valid} valid | {invalid} invalid | {missing} missing")
    print(f"{'-' * 70}")
    
    if invalid > 0:
        print("\nWARNING: Invalid signatures detected. Reports may have been modified.")
    elif valid > 0:
        print("\nOK: All signatures verified successfully. Reports integrity confirmed.")

print(f"\n{'=' * 70}")
````
